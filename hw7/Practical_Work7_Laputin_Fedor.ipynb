```python
from textblob import TextBlob


# Function to analyze sentiment
def analyze_sentiment(review):
    analysis = TextBlob(review)
    return analysis.sentiment.polarity, analysis.sentiment.subjectivity


# Apply the function to each DataFrame
hotel_reviews[["polarity", "subjectivity"]] = hotel_reviews["review"].apply(
    lambda x: pd.Series(analyze_sentiment(x))
)
cafe_reviews[["polarity", "subjectivity"]] = cafe_reviews["review"].apply(
    lambda x: pd.Series(analyze_sentiment(x))
)
restaurant_reviews[["polarity", "subjectivity"]] = restaurant_reviews["review"].apply(
    lambda x: pd.Series(analyze_sentiment(x))
)
```

```python
# Function to assign sentiment based on polarity
def get_sentiment(polarity):
    return 1 if polarity > 0 else 0


# Add sentiment column based on polarity
hotel_reviews["sentiment"] = hotel_reviews["polarity"].apply(get_sentiment)
cafe_reviews["sentiment"] = cafe_reviews["polarity"].apply(get_sentiment)
restaurant_reviews["sentiment"] = restaurant_reviews["polarity"].apply(get_sentiment)

# Display the updated DataFrames
hotel_reviews.head(20)
```

Ex.2. Training a Sentiment Model Using TFIDF and Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score

corpus = []
labels = []
for category, reviews in data.items():
    for review in reviews:
        corpus.append(review["review"])
        labels.append(1 if review["stars"] > 3 else 0)

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    corpus, labels, test_size=0.2, random_state=42
)

# Vectorize the text data using TFIDF
vectorizer = TfidfVectorizer(max_features=1000)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Train the logistic regression model
model = LogisticRegression()
model.fit(X_train_tfidf, y_train)

# Evaluate the model
y_pred_train = model.predict(X_train_tfidf)
train_accuracy = accuracy_score(y_train, y_pred_train)
train_f1 = f1_score(y_train, y_pred_train)
print("Train Accuracy:", train_accuracy)
print("Train F1:", train_f1)

y_pred_test = model.predict(X_test_tfidf)
test_accuracy = accuracy_score(y_test, y_pred_test)
test_f1 = f1_score(y_test, y_pred_test)
print("Test Accuracy:", test_accuracy)
print("Test F1:", test_f1)
```
